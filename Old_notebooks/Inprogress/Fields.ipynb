```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
from scipy.interpolate import griddata
from pylab import *
from numba import jit
import glob
```

```python
Comp="Mobilis"
OP="300daynokpp"
yrs='240-260'
x="/noc/msm/scratch/students/hb1g13/"+Comp+"/"+OP
fname=yrs+'all.nc'
os.chdir(x)
```

```python
file2read = netcdf.NetCDFFile(fname,'r')
U=file2read.variables['UVEL']
U=U[:]
Temp=file2read.variables['THETA']
temp=Temp[:]
Y=file2read.variables['Y']
Y=Y[:]
#os.chdir("/noc/msm/scratch/students/hb1g13/"+Comp+"/"+OP+"/grid/")
#file2read = netcdf.NetCDFFile('grid.nc','r')
#Z=file2read.variables['diag_levels']
#Z=Z[:]
```

```python
#os.chdir("/noc/msm/scratch/students/hb1g13/"+Comp+"/"+OP+"/grid/")
f = netcdf.NetCDFFile('grid.nc','r')
Z=f.variables['Z']
Z=Z[:]
```

```python
Utav=np.mean(U,axis=0)
Uzone=np.mean(Utav,axis=2)
plt.contourf(Y/1000,Z,Uzone,60,cmap=cm.jet)
clim((np.nanmin(Uzone)),np.nanmax(Uzone)-0.06)
cbar = plt.colorbar()
plt.title('Average U velocity')
plt.ylabel('Depth (m)')
plt.xlabel('Meridional Distance (km)')
cbar.ax.set_ylabel('Uvel (m/s)')
x="/noc/users/hb1g13/Figures/"+Comp+"/"+OP
os.chdir(x)
y="Ufield.png"
savefig(y)
```

```python
temptav=np.nanmean(temp,axis=0)
Tzone=np.nanmean(temptav,axis=2)
Tzone[Tzone==0]=np.nan
plt.contourf(Y/1000,Z,Tzone,60,cmap=cm.jet)
clim((np.nanmin(Tzone)),np.nanmax(Tzone))
cbar = plt.colorbar()
plt.title('Average Temperature')
plt.ylabel('Depth (m)')
plt.xlabel('Meridional Distance (km)')
cbar.ax.set_ylabel('Temp ($^o$C)')
x="/noc/users/hb1g13/Figures/"+Comp+"/"+OP
os.chdir(x)
y="Tempfield.png"
savefig(y)
```

```python
plt.contourf(Y/1000,Z,Uzone,60,cmap=cm.jet)
clim((np.nanmin(Uzone)),np.nanmax(Uzone)-0.06)
cbar = plt.colorbar()
plt.contour(Y/1000,Z,Tzone,4,colors='w',linewidths = 2)
plt.title('Average U velocity')
plt.ylabel('Depth (m)')
plt.xlabel('Meridional Distance (km)')
cbar.ax.set_ylabel('Uvel (m/s)')
x="/noc/users/hb1g13/Figures/"+Comp+"/"+OP
os.chdir(x)
y="UfieldwT.png"
savefig(y)
```
