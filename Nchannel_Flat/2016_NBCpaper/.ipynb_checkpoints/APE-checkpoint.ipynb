```python
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib import gridspec
from pylab import *
import sys
sys.path.append('/noc/users/hb1g13/Python/python_functions/')
import layers_calc_numba
import APE_Calc as AC
sys.path.append('/noc/users/hb1g13/Python/python_functions/MITgcmUtils/')
import utils
%matplotlib inline
```

```python
tau = ['50', '250', '500', '750', '1000', '2000', '3000', '10000']
Figletter = ['a) ', 'b) ', 'c) ', 'd) ', 'e)', 'f) ', 'g) ', 'h) ', 'j) ']
# Path root
x = '/noc/msm/scratch/students/hb1g13/Mobilis/Vary_N/'
# Now Make file structure
check = 0
runs = []
for i in range(len(tau)):
    flist = x+'/'+str(tau[i])+'m/VSQ.nc'
    if not os.path.exists(flist):
        print ' WARNING: '+flist+' does not exist! (skipping this tau...)'
        check += 0
    else:
        check += 1
        runs.append(i)
Runs = np.array(runs)
```

```python
fname = x+'/50m/'
c = utils.ChannelSetup(output_dir=str(fname))
```

```python
fig = plt.figure(figsize=(16.5, 16.5))
APEt = []
APEm = []
# Create Panels
Psi_levs = np.arange(-1.5, 4.15, .15)
Psi_ticks =np.arange(-.5, 4.15, .03)
Eddy_ticks = (np.arange(-3, 1, .5))
E_levs = np.arange(-3., 1, .25)
Q = c.mnc('SURF.nc', 'oceQnet').mean(axis=0).mean(axis=1)
gs = gridspec.GridSpec(5, 3, height_ratios=[1, 4, 4, 4, 4])
ax = plt.subplot(gs[0])
ax.plot(c.yc/1000,-Q, color='k', linewidth=2)
ax.set_title('Surface Heat Flux $W/m^2$', fontsize=20)
ax.set_ylabel('Heat Flux ($W/m^2$)', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax = plt.subplot(gs[1])
ax.plot(c.yc/1000,-Q, color='k', linewidth=2)
ax.set_title('Surface Heat Flux $W/m^2$', fontsize=20)
ax.set_ylabel('Heat Flux ($W/m^2$)', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax = plt.subplot(gs[2])
ax.plot(c.yc/1000,-Q, color='k', linewidth=2)
ax.set_title('Surface Heat Flux $W/m^2$', fontsize=20)
ax.set_ylabel('Heat Flux ($W/m^2$)', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
for i in range(len(Runs)):
    
    fname = x+'/'+str(tau[Runs[i]])+'m/'
    c = utils.ChannelSetup(output_dir=str(fname))
    # PLOT PANNELS
    g = AC.APEcomputer(c)
    APE = g.calc_APE()
    
    APEt.append(np.sum(abs(APE)))
    APEm.append(np.mean(APE))
    ax = plt.subplot(gs[3+i])
    Q2_levs = (np.arange(-1,8,1))
    p = ax.contourf(c.xc/1000, c.yc/1000,
                (APE)/(1000*c.H*5000*5000), cmap=plt.cm.plasma_r, extend='both')

    ax.set_title(str(Figletter[Runs[i]])+str(tau[Runs[i]])+' m', fontsize=30)
    ax.set_xlabel('Distance (km)', fontsize=30)
    ax.set_ylabel('Depth (km)', fontsize=30)
    ax.tick_params(axis='both', which='major', labelsize=30)
plt.tight_layout()
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbar = fig.colorbar(p, cax=cax)
cbar.ax.set_ylabel('APE $(m^2s^{-2})$', fontsize=30, )
cbar.ax.tick_params(labelsize=30)
```

```python
plt.plot([50,250,500,1000,2000,3000,10000],np.array(APEt)*10**-15,'-ko')
plt.xscale("log")
plt.xlabel('Relaxation timescale (days)', fontsize=20)
plt.ylabel(r'PJ', fontsize=20)
plt.title('Domain integrated APE ', fontsize=20)
#plt.ylim(0,4)
plt.xlim(0,2*10**4)
```

```python
fig = plt.figure(figsize=(20.5, 24.5))
# Create Panels
Psi_levs = np.arange(-5.5, 5.5, .05)#+.5
Psi_ticks =np.arange(-.5, 4.15, .03)
Eddy_ticks = (np.arange(-3, 1, .5))
E_levs = np.arange(-3., 1, .25)
Q = c.mnc('SURF.nc', 'oceQnet').mean(axis=0).mean(axis=1)
gs = gridspec.GridSpec(5, 3, height_ratios=[1, 4, 4, 4, 4])
ax = plt.subplot(gs[0])
ax.plot(c.yc/1000,-Q, color='k', linewidth=2)
ax.set_title('Surface Heat Flux $W/m^2$', fontsize=20)
ax.set_ylabel('Heat Flux ($W/m^2$)', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax = plt.subplot(gs[1])
ax.plot(c.yc/1000,-Q, color='k', linewidth=2)
ax.set_title('Surface Heat Flux $W/m^2$', fontsize=20)
ax.set_ylabel('Heat Flux ($W/m^2$)', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax = plt.subplot(gs[2])
ax.plot(c.yc/1000,-Q, color='k', linewidth=2)
ax.set_title('Surface Heat Flux $W/m^2$', fontsize=20)
ax.set_ylabel('Heat Flux ($W/m^2$)', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
APEdeep =[]
APEnorth = []
APEsouth = []
for i in range(len(Runs)):
    
    fname = x+'/'+str(tau[Runs[i]])+'m/'
    c = utils.ChannelSetup(output_dir=str(fname))
    # PLOT PANNELS
    g = AC.APEcomputer(c)
    APE = g.calc_APE_3d().mean(axis=2)
    APEdeep.append(np.max((APE[28::,:])))
    APEnorth.append(np.max((APE[0:28,:])))
    APEsouth.append(np.min((APE[:,:])))
    ax = plt.subplot(gs[3+i])
    Q2_levs = (np.arange(-1,8,1))
    p = ax.contourf(c.yc/1000, c.zc/1000,
                (APE)/(1000*c.H*5000*5000),Psi_levs*10**-4, cmap=plt.cm.seismic, extend='both')
    ax.set_title(str(Figletter[Runs[i]])+str(tau[Runs[i]])+' m', fontsize=30)
    ax.set_xlabel('Distance (km)', fontsize=30)
    ax.set_ylabel('Depth (km)', fontsize=30)
    ax.tick_params(axis='both', which='major', labelsize=30)
plt.tight_layout()
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbar = fig.colorbar(p, cax=cax)
cbar.ax.set_ylabel('APE $(m^2s^{-2})$', fontsize=30, )
cbar.ax.tick_params(labelsize=30)
APEdeep = np.array(APEdeep)
APEnorth = np.array(APEnorth)
APEsouth = np.array(APEsouth)
```
