```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
import glob
import matplotlib.gridspec as gridspec
```

```python
# Define a class that forces representation of float to look a certain way
# This remove trailing zero so '1.0' becomes '1'
class nf(float):
     def __repr__(self):
         str = '%.1f' % (self.__float__(),)
         if str[-1]=='0':
             return '%.0f' % self.__float__()
         else:
             return '%.1f' % self.__float__()
```

```python
# Label levels with specially formatted floats
if plt.rcParams["text.usetex"]:
    fmt = r'%r $^oC$ %'
else:
    fmt = '%r $^oC$'
```

```python
tau=['3','10','30','100','300','1000','3000','10000','Closed']
x='/noc/msm/scratch/students/hb1g13/Mobilis'
```

```python
check=0
runs=[]
for i in range(len(tau)):
    flist=x+'/'+str(tau[i])+'daynokpp/VV.nc'
    if not os.path.exists(flist):
       print 'WARNING: '+flist+' does not exist! (skipping this ta\
u...)'
       check+=0
    else:
       check+=1
       runs.append(i)
```

```python
Runs=np.array(runs)
```

```python
filename=x+'/'+str(tau[Runs[0]])+'daynokpp/grid.nc'
file2read = netcdf.NetCDFFile(filename,'r')
Y=file2read.variables['Y']
Y=Y[:]*1
Yp=file2read.variables['Yp1']
Yp=Yp[:]*1
z=file2read.variables['Z']
Z=z[:]*1
Zp1=file2read.variables['Zp1']
Zp=Zp1[:]*1
dz=Zp[0:len(Zp)-1]-Zp[1:len(Zp)]
dA=dz*5000
```

```python
fig=plt.figure(figsize=(10.5,12.5))
EKE=np.zeros(np.shape(Runs))
for i in range(len(Runs)):
    fname=x+'/'+str(tau[Runs[i]])+'daynokpp/VV.nc'
    file2read = netcdf.NetCDFFile(fname,'r')
    VV=file2read.variables['VV']
    VVzone=0.5*np.mean(VV[:]*1,axis=2)
    EKE[i]=np.mean(VVzone)
    filename=x+'/'+str(tau[Runs[i]])+'daynokpp/Tav.nc'
    file2read = netcdf.NetCDFFile(filename,'r')
    Temp=file2read.variables["THETA"] 
    Temp=Temp[:]
    Tzone=np.mean(np.squeeze(Temp*1),axis = 2)
    Q2_levs = (np.arange(0,8,1))
    y=Y/1000
    Q_levs = np.arange(-0.1,0.1,0.005)
    Q_ticks = np.arange(-0.1,0.1,0.025)
    ax = fig.add_subplot(3, 3, i+1)
    p=ax.contourf(Yp/1000,Z,VVzone,Q_levs,cmap=plt.cm.seismic) #Use b2r colourmap
    q=ax.contour(Y/1000,Z,Tzone,Q2_levs,colors='k',linewidths = 2)
    q.levels = [nf(val) for val in q.levels ]
    plt.clabel(q,q.levels[::2], inline=1,fmt=fmt, fontsize=10)
    #clim(-0.1,0.1) # Put 0 to white
    #cbar=plt.colorbar(cf, ticks=Psi_ticks, shrink=0.8)
    ax.set_title(str(tau[Runs[i]])+'day')
    if str(tau[Runs[i]])=='Closed':
        ax.set_title(str(tau[Runs[i]]))
    ax.set_xlabel('Distance (km)')
    ax.set_ylabel('Depth (m)')
    #cbar=plt.colorbar(p, ticks=Psi_ticks, shrink=0.8)
    #ax.set_clim(-1,1)
    #cbar.ax.set_ylabel('$\psi \,\, (sv)$')
plt.tight_layout()
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbar=fig.colorbar(p, cax=cax, ticks=Q_ticks)
cbar.ax.set_ylabel('EKE')
q=( os.path.expanduser('~')+"/Figures/Mobilis")
if not os.path.exists(q):
       os.makedirs(q)
plt.savefig(q+"/EKEpanel.png")
```

```python
plt.plot(['3','10','30','100','300','1000','3000','10000'],EKE[0:-1]*1000,'o')
plt.xscale('log')
plt.ylabel('EKE x10$^-3$')
plt.xlabel('Relaxation timescale')
```

That's probably the best clear cut relationship for what changes when you change the NBC!

Changing Stratification at NBC must change APE? Lets look at that? 

\begin{equation}
APE=-\int \frac{b^2}{2N^2}\mathrm{d}x
\end{equation}

\begin{equation}
N=\sqrt{\frac{g}{\rho}\frac{d\rho}{dz}}
\end{equation}

```python
G=-9.81
alpha=2.000000E-04
N=0.003
```

```python
fig=plt.figure(figsize=(10.5,12.5))
EKE=np.zeros(np.shape(Runs))
APE=np.zeros(np.shape(Runs))
for i in range(len(Runs)):
    filename=x+'/'+str(tau[Runs[i]])+'daynokpp/Tav.nc'
    file2read = netcdf.NetCDFFile(filename,'r')
    Temp=file2read.variables["THETA"] 
    Temp=Temp[:]
    Tzone=np.mean(np.squeeze(Temp*1),axis = 2)
    B2=(-G*alpha*(Tzone-3.0))**2
    APE[i]=np.mean(B2/(2*N**2))
    Q2_levs = (np.arange(0,8,1))
    y=Y/1000
    Q_levs = np.arange(-5,5,0.25)
    Q_ticks = np.arange(-5,5,0.5)
    ax = fig.add_subplot(3, 3, i+1)
    p=ax.contourf(Y/1000,Z,B2/(2*N**2),Q_levs,cmap=plt.cm.seismic) #Use b2r colourmap
    q=ax.contour(Y/1000,Z,Tzone,Q2_levs,colors='k',linewidths = 2)
    q.levels = [nf(val) for val in q.levels ]
    plt.clabel(q,q.levels[::2], inline=1,fmt=fmt, fontsize=10)
    #clim(-0.1,0.1) # Put 0 to white
    #cbar=plt.colorbar(cf, ticks=Psi_ticks, shrink=0.8)
    ax.set_title(str(tau[Runs[i]])+'day')
    if str(tau[Runs[i]])=='Closed':
        ax.set_title(str(tau[Runs[i]]))
    ax.set_xlabel('Distance (km)')
    ax.set_ylabel('Depth (m)')
    #cbar=plt.colorbar(p, ticks=Psi_ticks, shrink=0.8)
    #ax.set_clim(-1,1)
    #cbar.ax.set_ylabel('$\psi \,\, (sv)$')
plt.tight_layout()
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbar=fig.colorbar(p, cax=cax, ticks=Q_ticks)
cbar.ax.set_ylabel('EKE')
q=( os.path.expanduser('~')+"/Figures/Mobilis")
if not os.path.exists(q):
       os.makedirs(q)
plt.savefig(q+"/APEpanel.png")
```
