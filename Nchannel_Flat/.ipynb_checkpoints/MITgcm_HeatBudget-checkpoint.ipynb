#  <center> Heat Budget </center> #

(1) implicitFreeSurface=.TRUE., (default configuration)

(2) nonlinFreeSurf=0, (default) 

**Heat budget at a particular point (ix,iy,iz):**

CellVol = RAC(ix,iy) * DRF(iz) * hFacC(ix,iy,iz)

Tendency due to advection:

**Adv_tend** = - [ {ADVr_TH(ix,iy,iz) – ADVr_TH(ix,iy,iz+1)}/CellVol +
 {ADVx_TH(ix+1,iy,iz) – ADVx_TH(ix,iy,iz)}/CellVol +
 {ADVy_TH(ix,iy+1,iz) – ADVy_TH(ix,iy,iz)}/CellVol ] 


*Tendency due to diffusion = 0*

Tendency at surface due to incident heat flux:

**Tflx_tend** = (TFLUX(ix,iy))) / (rhoConst * Cp * DRF(1) * hfacC(ix,iy,iz)) 

Tendency of mass correction at the surface (due to Linear Free Surface approximation): 

**Surf_corr_tend** = (tsurfcor- WTHMASS(ix,iy)) / (DRF(1) * hFacC(ix,iy,1))

Where tsurfcor = SUM(WTHMASS * RAC) / globalArea (if using linFSConserveTr = True, and
= 0 otherwise.)


***

Thus heat budget at surface (iz = 1):

**surf_heat_tend = Adv_tend + Dif_tend + Kpp_tend + Qsw_tend + Tflx_tend + Surf_corr_tend**

Heat budget at subsurface (iz>1):

**sub_surf_heat_tend = Adv_tend + Dif_tend + Kpp_tend + Qsw_tend**

These heat budget tendencies should be equal to **TOTTTEND(ix,iy,iz)/86400.0** (for linear free
surface).

```python
fig = plt.figure(figsize=(16.5, 16.5))
for i in range(len(Runs)):
    #i=0
    fname = x+'/'+str(tau[Runs[i]])+'daynokpp/'
    c = utils.ChannelSetup(output_dir=str(fname))
    # PLOT PANNELS
    Nx,Ny,Nz = c.Nx, c.Ny, c.Nz
    # load V, W, T bar                                                                           
    # put everything on the C Grid                                                          
    T = c.mnc('Tav.nc','THETA')
    # Tendency of Potential Temperature degC/day
    TEND = c.mnc('HEAT.nc','TOTTTEND')/86400
    # Meridional Advective Flux of Pot.Temperature degC.m^3/s
    ADVy_TH = (c.mnc('HEAT.nc','ADVy_TH'))
    # Vertical   Advective Flux of Pot.Temperature degC.m^3/s
    ADVr_TH = c.mnc('HEAT.nc','ADVr_TH')
    npad = ((0, 1), (0, 0), (0,0))
    ADVr_TH = np.pad(ADVr_TH, pad_width=npad, mode='constant', constant_values=0)
    # Zona   Advective Flux of Pot.Temperature degC.m^3/s
    ADVx_TH = c.mnc('HEAT.nc','ADVx_TH')
    Adv_tend = zeros((c.Nz,c.Ny,c.Nx))
    for ix in range(c.Nx-1):
        for iy in range(c.Ny-1):
            for iz in range(c.Nz-1):
                Adv_tend[iz,iy,ix] = - ( (ADVr_TH[iz,iy,ix] - ADVr_TH[iz+1,iy,ix])/CellVol[iz,iy,ix] + 
                               (ADVx_TH[iz,iy,ix+1] - ADVx_TH[iz,iy,ix])/CellVol[iz,iy,ix] + 
                               (ADVy_TH[iz,iy+1,ix] - ADVy_TH[iz,iy,ix])/CellVol[iz,iy,ix] )
    # Vertical Mass-Weight Transp of Pot Temp (K.m/s)
    WTHMASS = c.mnc('SURF.nc', 'WTHMASS')
    # Surface cor
    Surcor = (WTHMASS.mean(axis=0)/c.dzf[0]).mean(axis=1)
    #tcor = sum(WTHMASS.mean(axis=0)/CellVol[0,:,:])#.sum(axis=1)
    #Surcor = (-WTHMASS.mean(axis=0)/CellVol[0,:,:]).sum(axis=1)
    # ADV
    HeatDiv = -sum(Adv_tend[:,,100],axis=0)
    # total heat flux (match heat-content variations), >0 increases theta W/m^2
    TFLUX =  c.mnc('SURF.nc', 'TFLUX')  
    Tflux = TFLUX.mean(axis=2).mean(axis=0)/(c.dzf[0]*1000*3985)
    # Tendancy total
    Tend = TEND.mean(axis=2).sum(axis=0)
    ax = fig.add_subplot(2, 2, i+1)
    p = plt.plot(c.yc/1000,Tflux*10**6,'k',linewidth=3)
    p = plt.plot(c.yc/1000,HeatDiv*10**6,'r',linewidth=2)
    #p = plt.plot(c.yc/1000,tcor*10**6,'--c',linewidth=1)
    p = plt.plot(c.yc/1000,Surcor*10**6,'c',linewidth=1)
    p = plt.plot(c.yc/1000,Tend*10**6,'--k',linewidth=2)
    plt.ylim(-1,1)
    ax.set_title(str(Figletter[Runs[i]])+str(tau[Runs[i]])+'day', fontsize=30)
    if str(tau[Runs[i]]) == 'Closed':
        ax.set_title(str(Figletter[Runs[i]])+str(tau[Runs[i]]), fontsize=30)
    ax.set_xlabel('Distance (km)', fontsize=30)
    ax.set_ylabel(r'$^{o}C /s$', fontsize=30)
    ax.tick_params(axis='both', which='major', labelsize=30)
lgd=plt.legend([r'Q','Tflux','TEND'],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,fontsize=20)
plt.tight_layout()
```
