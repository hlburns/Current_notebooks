**Relating the ROC to NBC**

The ROC depends on the MOC and an opposing eddy component. The eddy component is often written as:
\begin{equation}
\psi ^* = KSp,
\end{equation}
where the eddy overturning ($\psi ^*$) depends on eddy diffusivity (K) and isopycnal slope (Sp). The eddy diffusivity depends on a number of factors including the background vertical buoyancy gradient (Ferreira,Marshall, Hiembach 2005} which is influenced by the imposed buoyancy at the Northern boundary. Although the mean component can be calculated from wind forcing along the eddy component relies on other factors implying the ROC can not be determined from local surface forcing alone:
\begin{equation}
\psi ^* = -\frac{K_{ref}}{N^2_{ref}}\frac{\partial b}{\partial y},
\end{equation} 
so
\begin{equation}
\psi_{res}= \underbrace{\frac{\tau _s}{\rho _0 f}}_\text{wind} + \underbrace{\frac{K_{ref}}{N^2_{ref}}}_\text{Set by NBC} \frac{\partial b}{\partial y},
\end{equation}
or explicitly with northern boundary relation:
\begin{equation}
\psi_{res}= \frac{\tau _s}{\rho _0 f} -S_p \frac{K_{ref}}{N^2_{ref}} \frac{\partial b_N}{\partial z}.
\end{equation}
