```python
# Grid parameters
gridfilepath=path+'/'+str(tau[0])+'daynokpp/'
file2=netCDF4.Dataset(gridfilepath+'grid.nc','r')
Zp=file2.variables['Zp1'][:]
Z=file2.variables['Z'][:]
Y=file2.variables['Yp1'][:]
Yc=file2.variables['Y'][:]
dz=Zp[0:-1]-Zp[1::][:]
dA=dz*5000
Zmatrix=np.zeros((len(Z),len(Yc)))
Zmatrix[:,:]=1
dx=Y[1]-Y[0] # Find Resolution
Zmatrix=np.apply_along_axis(np.multiply,0,Zmatrix,dz)
```

```python
fig=plt.figure(figsize=(16.5,16.5))
for i in range(len(Runs)):
    A=path+'/'+str(tau[Runs[i]])+'daynokpp/Tav.nc'
    file2 = netCDF4.Dataset(A,'r')
    Temp=file2.variables['THETA'][:]
    Tav=np.mean(np.mean(Temp,axis=0),axis=2)
    bz, by = np.gradient(Tav)
    by = by/5000
    bz = bz/Zmatrix
    by[by<0.0015/10000]=0.0015/10000
    bz[bz>-0.002/10]=-0.002/10
    S= by/bz
    S_levs= np.arange(-.13,.0005,.01)/30
    if Full=='N':
         ax = fig.add_subplot(2, 2, i+1)
         p2=ax.contourf(Yc/1000,Z,S,S_levs)
         
    if str(tau[Runs[i]])=='Closed':
        ax.set_title(str(tau[Runs[i]]),fontsize=20)
    ax.set_xlabel('Distance (km)',fontsize=20)
    ax.set_ylabel('$m^2s^{-3}$',fontsize=20)
    ax.tick_params(labelsize=30)
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbar=fig.colorbar(p2, cax=cax)
plt.tight_layout()
```
