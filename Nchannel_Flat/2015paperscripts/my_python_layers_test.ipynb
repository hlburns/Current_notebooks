```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('/noc/users/hb1g13/Python/python_functions/')
import layers_calc_t
sys.path.append('/noc/users/hb1g13/Python/python_functions/MITgcmUtils/')
import utils
%matplotlib inline
```

```python
# Set up Figure:
psi_levs = np.arange(-2., 2., .21)
psi_ticks = np.arange(-2., 2., .5)
fig = plt.figure(figsize=(20.0, 10.0))
# Instantaneous folder:
fname = '/noc/msm/scratch/students/hb1g13/Mobilis/instantaneous/'
c = utils.ChannelSetup(output_dir=str(fname))
g = layers_calc_t.LayersComputer(c)
# MITgcm layer intergrated transport:
lvrho = c.mnc_t('PSI.nc', 'LaVH1TH', 0)
V = lvrho.mean(axis=2)
psi = (-V.cumsum(axis=0)*c.Lx/10**6)
# Plot it:
y = c.yg/1000
Rho = g.glvl
ax = fig.add_subplot(1, 2, 1)
p = ax.contourf(y, Rho, psi, psi_levs, cmap=plt.cm.seismic, extend='both')
ax.set_title('a) MIT layers package ROC', fontsize=40)
ax.set_xlabel('Distance (km)', fontsize=40)
ax.set_ylabel(r'Temperature $^o$C', fontsize=40)
ax.tick_params(axis='both', which='major', labelsize=40)
# My Python Layer intergrated transport:
lvrho, th = g.compute_vflux() # See layers_calc_t.py for the 200 lines of code req.
# Turn it to a ROC in temp space!
V = lvrho.mean(axis=2)
psi = (-V.cumsum(axis=0)*c.Lx/10**6)
# Plot it
y = c.yc/1000
ax = fig.add_subplot(1, 2, 2)
p = ax.contourf(y, Rho, psi, psi_levs, cmap=plt.cm.seismic, extend='both')
ax.set_title('b) My python layers ROC', fontsize=40)
ax.set_xlabel('Distance (km)', fontsize=40)
ax.set_ylabel(r'Temperature $^o$C', fontsize=40)
plt.tight_layout()
# Add Colorbar
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbar = fig.colorbar(p, cax=cax, ticks=psi_ticks)
cbar.ax.set_ylabel(r'$\Psi \,\, (sv)$', fontsize=40)
cbar.ax.tick_params(labelsize=40)
```
