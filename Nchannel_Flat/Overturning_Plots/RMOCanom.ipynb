```python
from scipy.io import netcdf
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
import glob
import matplotlib.gridspec as gridspec
```

```python
Year='240-260Psi.nc'
```

```python
#tau=['3','10','30','100','300','1000','3000','10000','Closed']
tau=['3','30','300','3000','10000','Closed']
x='/noc/msm/scratch/students/hb1g13/Mobilis'
```

```python
check=0
runs=[]
for i in range(len(tau)):
    flist=x+'/'+str(tau[i])+'daynokpp/'+Year
    if not os.path.exists(flist):
       print 'WARNING: '+flist+' does not exist! (skipping this ta\
u...)'
       check+=0
    else:
       check+=1
       runs.append(i)
```

```python
Runs=np.array(runs)
```

```python
filename=x+'/'+str(tau[Runs[0]])+'daynokpp/'+Year
file2read = netcdf.NetCDFFile(filename,'r')
lvrho=file2read.variables["LaVH1TH"]
lvrho=lvrho[:]
Y=file2read.variables['Yp1']
Y=Y[:]
dx=Y[1]-Y[0] # Find Resolution
VT=np.sum(lvrho*dx,axis=3) #integrate Vdx along x
VTfdz=np.cumsum(VT[:,::-1,:],axis=1) #sum up the water column
psi1=np.mean(VTfdz[:,::-1,:],axis=0)/10**6 #Time average and pu\
```

```python
fig=plt.figure(figsize=(20.0,15.0))
for i in range(len(Runs)):
    fname=x+'/'+str(tau[Runs[i]])+'daynokpp/'+Year
    file2read = netcdf.NetCDFFile(fname,'r')
    lvrho=file2read.variables["LaVH1TH"]
    lvrho=lvrho[:]
    Y=file2read.variables['Yp1']
    Y=Y[:]
    time=file2read.variables['T']
    ti=time[:]
    dx=Y[1]-Y[0] # Find Resolution
    VT=np.sum(lvrho*dx,axis=3) #integrate Vdx along x
    VTfdz=np.cumsum(VT[:,::-1,:],axis=1) #sum up the water column
    psi=np.mean(VTfdz[:,::-1,:],axis=0)/10**6 #Time average and pu\
#t into Sv and put back in right order
    y=Y/1000
    z=np.array(range(1,60))
    psianom=(psi)-(psi1)
    Rho = np.genfromtxt(x+'/'+str(tau[Runs[i]])+'daynokpp/Temp', delimiter = ',')
    nolayers=len(psi[:,1])
    Rho=Rho[0:nolayers]#The layers package bins a layer so adjust \
    start=int(np.divide(ti[0],(86400*360)))#Find run start and sto\
    end=int(np.divide(ti[-1],(86400*360)))
    Q_levs = (np.arange(-30,30)+0.5)
    Psi_levs = Q_levs / 10
    Q_ticks = np.arange(-16,20,4.)
    Psi_ticks = Q_ticks / 10
    #if np.max(abs(psi))>1:
    #   Psi_levs = (np.arange(-30,30)+0.5)/10
    ax = fig.add_subplot(2, 3, i+1)
    #CF=plt.subplot(3,2,i+1)
    psianom[psianom>2.5]=2.5 #Force scale
    p=ax.contourf(y,Rho,psianom,Psi_levs,cmap=plt.cm.seismic) #Use b2r colourmap
    #clim(-1,1) # Put 0 to white
    #cbar=plt.colorbar(cf, ticks=Psi_ticks, shrink=0.8)
    ax.set_title(str(tau[Runs[i]])+'day',fontsize=20)
    ax.tick_params(axis='both', which='major', labelsize=20)
    if str(tau[Runs[i]])=='Closed':
        ax.set_title(str(tau[Runs[i]]),fontsize=20)
    ax.set_xlabel('Distance (km)',fontsize=20)
    ax.set_ylabel('Temperature $^o$C',fontsize=20)
    #cbar=plt.colorbar(p, ticks=Psi_ticks, shrink=0.8)
    #ax.set_clim(-1,1)
    #cbar.ax.set_ylabel('$\psi \,\, (sv)$')
plt.tight_layout()
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbar=fig.colorbar(p, cax=cax,ticks=Psi_ticks)
cbar.ax.set_ylabel('$\psi \,\, (sv)$',fontsize=30)
cbar.ax.tick_params(labelsize=30)
q=( os.path.expanduser('~')+"/Figures/Mobilis")
if not os.path.exists(q):
       os.makedirs(q)
plt.savefig(q+"/Psianompanel"+str(start)+"-"+str(end)+".png")
```
