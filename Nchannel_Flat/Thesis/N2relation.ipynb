```python
# Path root
x = '/noc/msm/scratch/students/hb1g13/Mobilis/'
# Now Make file structure

tau = [ 'deltat/4dt/','deltat/8dt/','Vary_N/500m/','Vary_N/250m/','Vary_N/750m/','Vary_N/1000m/','Vary_N/2000m/', 'Vary_N/3000m/', 'Vary_N/10000m/', 
       'deltat/16dt/' , '3daynokpp/', '30daynokpp/',  '300daynokpp/',  '3000daynokpp/',
       '10daynokpp/', '100daynokpp/',  '10000daynokpp/',
        'Closeddaynokpp/']

check = 0
runs = []
for i in range(len(tau)):
    flist = x+'/'+str(tau[i])+'VSQ.nc'
    if not os.path.exists(flist):
        print ' WARNING: '+flist+' does not exist! (skipping this tau...)'
        check += 0
    else:
        check += 1
        runs.append(i)
Runse=np.array(runs)
Runs=np.array(tau)
Figletter = ['a) ','b) ','c) ','d) ','e)','f)','g)','h)','j)']
```

```python
# Loop through runs
ROC = []
N2t = []
for i in range(len(Runse)):
    fname = fname = x+str(tau[Runse[i]])
    c = utils.ChannelSetup(output_dir=str(fname))
    CellVol = 5000*np.tile(c.dzf,(c.Ny,1)).T
    # Layers package variables
    Psi = c.get_psi_iso_z()/10**6

    # Make list of ROC / ROC0
    if i == 0:
        ROC0 = np.sum(abs(Psi))
    ROC.append((np.sum(abs(Psi))-ROC0)/ROC0)
    N2 = c.get_N2()[-30::]
    if i == 0:
        N20 = np.sum(abs(N2*CellVol))/(c.H*c.Ly)
    N2t.append(((np.sum(abs(N2*CellVol)))/(c.H*c.Ly)))
```

```python
fig, ax = plt.subplots(figsize=(7.5,7.5))
X =np.log((np.array(N2t)-np.array(N2t).min())*100)
X=X[1::]
ax.plot(X, np.array(ROC)[1::]*100,'rx', linewidth=2)
#plt.xscale('log', basex=e)
plt.xlabel('$N^2$ % increase', fontsize=20)
plt.ylabel('% change in ROC', fontsize=20)
plt.title("% change in ROC with increasing $N^2$ \n",fontsize=30)
#plt.xticks(np.arange(0,5,1),fontsize=18)
plt.yticks(fontsize=25)
a =ax.get_xticks().tolist()
#a = ['0','$e^0$','$e^1$','$e^2$','$e^3$','$e^4$']
def ticks(y, pos):
    if pos ==0:
        return r'0'.format(np.log(y))
    else:
        return r'$ln{:.0f}$'.format(y)
ax.xaxis.set_major_formatter(mtick.FuncFormatter(ticks))
m, c, r, p, ste = linregress(X ,np.array(ROC)[1::]*100) 
Y = m*np.arange(0,4,0.1)+c
q = plt.plot(np.arange(0,4,0.1), Y,'k', linewidth=2)
plt.yticks(fontsize=20)
print('m = ', m, 'c = ', c, 'R = ', r , 'P = ', p , 'ste =', ste)
```

```python
# Loop through runs
ROCt = []
N2t = []
EKEt = []
APEt = []
APEdt = []
APEst = []
APEnt = []
Dsum = []
for i in range(len(Runse)):
    fname = x+str(tau[Runse[i]])
    c = utils.ChannelSetup(output_dir=str(fname))
    CellVol = 5000*np.tile(c.dzf,(c.Ny,1)).T
    # Layers package variables
    Psi = c.get_psi_iso_z()/10**6
    EKE = c.calc_EKE()*CellVol
    
   
    # Make list of ROC / ROC0
    if i == 0:
        ROC0 = np.sum(abs(Psi))
        EKE0 = np.sum(abs(EKE))/(c.H*c.Ly)
        
    ROCt.append((np.sum(abs(Psi))-ROC0)/ROC0)
    EKEt.append(((np.sum(abs(EKE)))/(c.H*c.Ly)-EKE0)/EKE0)
    
    N2 = c.get_N2()
    if i == 0:
        N20 = np.sum(abs(N2*CellVol))/(c.H)
    N2t.append(((np.sum(abs(N2*CellVol)))/(c.H)-N20)/N20)

EKEt = np.array(EKEt)
N2 = np.array(N2)
ROC = np.array(ROC)
```

```python
fig, ax = plt.subplots(figsize=(7.5,7.5))
ax.plot(np.array(N2t)*100, EKEt*100,'rx', linewidth=2)
plt.xlabel('$N^2$ % increase', fontsize=20)
plt.ylabel('% increase in EKE', fontsize=20)
plt.title("% increase in EKE magintude with increasing $N^2$ \n",fontsize=30)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
m, c, r, p, ste = linregress(np.array(N2t)*100 ,EKEt*100) 
Y = m*np.arange(0,30,1)+c
q = plt.plot(np.arange(0,30,1), Y,'k', linewidth=2)
plt.xscale('linear')
print('m = ', m, 'c = ', c, 'R = ', r , 'P = ', p , 'ste =', ste)
```
