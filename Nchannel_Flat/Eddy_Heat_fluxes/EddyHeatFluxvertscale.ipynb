```python
mycols=['b','g','k','c','m','y','k','r','g']
linestyles=['-','-','-','-','-','-','-','--','--']
```

```python
fname=x+'/'+str(tau[Runs[0]])+'daynokpp/grid.nc'
file2=netcdf.netcdf_file(fname,'r')
Zp1=file2.variables['Zp1']
Zp=Zp1[:]*1
Z=file2.variables['Z']
Z=Z[:]*1
dx=5000
dz=Zp[0:len(Zp)-1]-Zp[1:len(Zp)]
dA=dz*5000
```

```python
for i in range(len(Runs)):
    fname=x+'/'+str(tau[Runs[i]])+'daynokpp/VTprimebar.nc'
    file2read = netcdf.NetCDFFile(fname,'r')
    VTbar=file2read.variables['VT']
    VTdA=np.zeros(shape(VTbar))
    for k in range(len(dz)):
        VTdA[k,:,:]=VTbar[k,:,:]*dA[k]
    EddyHF=1030*3985*(np.sum(np.sum(VTdA,axis=1),axis=1))/10**15
    #EddyHF0=1030*3985*(np.sum(np.sum(VTdA[0:3,:,:],axis=0),axis=1))/10**15
    E0=plt.plot(EddyHF,Z,color=mycols[i],linestyle=linestyles[i])
    #cbar=plt.colorbar(p, ticks=Psi_ticks, shrink=0.8)
    #ax.set_clim(-1,1)
    #cbar.ax.set_ylabel('$\psi \,\, (sv)$')
plt.title("Meridional Eddy Fluxes")
plt.xlabel("Heat Flux (PW)")
plt.ylabel("Depth (m)")
runnames=[]
for i in range(len(Runs)):
    runnames.append(tau[Runs[i]])
lgd=plt.legend(runnames, shadow=True,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
q=( os.path.expanduser('~')+"/Figures/Mobilis")
if not os.path.exists(q):
       os.makedirs(q)
plt.savefig(q+"/SurfaceEddy.png",bbox_extra_artists=(lgd,), bbox_inches='tight')
```

```python
#Regrid V to center points
def regrid(Variable):
    Vc=(Variable[:,:,0:-1,:]+Variable[:,:,1::,:])/2
    return Vc
numba_regrid = autojit()(regrid)
numba_regrid.func_name = "numba_regrid"
```

```python
for i in range(len(Runs)):
    fname=x+'/'+str(tau[Runs[i]])+'daynokpp/VTprimebar.nc'
    fname2=x+'/'+str(tau[Runs[i]])+'daynokpp/Tav.nc'
    fname3=x+'/'+str(tau[Runs[i]])+'daynokpp/VTbar.nc'
    file2 = netcdf.NetCDFFile(fname2,'r') 
    V=file2.variables['VVEL']
    VC=numba_regrid(V[:]*1)
    Temp=file2.variables['THETA']
    VTAVC=np.squeeze(VC)
    TTAV=np.squeeze(Temp[:]*1)
    file2read = netcdf.NetCDFFile(fname,'r')
    VTbar=file2read.variables['VT']
    VTdA=np.zeros(shape(VTbar))
    for k in range(len(dz)):
        VTdA[k,:,:]=VTbar[k,:,:]*dA[k]
    EddyHF=1030*3985*(np.sum(VTdA,axis=2))/10**15
    VT=VTAVC*TTAV
    VTdA=np.zeros(shape(VT))
    for k in range(len(dz)):
        VTdA[k,:,:]=VT[k,:,:]*dA[k]
    MeanHF=1030*3985*(np.sum(VTdA,axis=2))/10**15
    fileVT = netcdf.NetCDFFile(fname3,'r') 
    VTbar=fileVT.variables['VT']
    VTdA=np.zeros(shape(VT))
    for k in range(len(dz)):
        VTdA[k,:,:]=VTbar[k,:,:]*dA[k]
    TotalHF=1030*3985*(np.sum(np.sum(VTdA,axis=1),axis=1))/10**15
    E0=plt.plot(np.min(EddyHF,axis=1),Z,color=mycols[i],linestyle=linestyles[i])
    #T0=plt.plot(TotalHF,Z,'k',linewidth=2)
    M0=plt.plot(np.max(MeanHF,axis=1),Z,'r')
```
