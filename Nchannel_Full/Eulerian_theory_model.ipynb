```python
import netCDF4
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import sys
import glob
import math
%matplotlib inline
```

```python
#path = '/hpcdata/scratch/hb1g13/NchannelFlat'
path = '/noc/msm/scratch/students/hb1g13/Mobilis/Full'
```

```python
gridfilepath=path+'/3/'
file2 = netCDF4.Dataset(gridfilepath+'grid.nc', 'r')
Zp = file2.variables['Zp1'][:]
Z = file2.variables['Z'][:]
Yp = file2.variables['Yp1'][:]
Yc = file2.variables['Y'][:]
X = file2.variables['X'][:]
Zmatrix = np.zeros((len(Z), len(Yp)))
Zmatrix[:, :] = 1
dx = Yc[1]-Yc[0]  # Find Resolution
dz = Zp[0:len(Zp)-1]-Zp[1:len(Zp)]
Vmask = file2.variables['HFacS'][:]
Tmask = file2.variables['HFacC'][:]
Zmatrix = np.apply_along_axis(np.multiply, 0, Zmatrix, dz)
X=file2.variables['Xp1'][:]
```

```python
# Constant Parameters
alpha = 2.000000E-04
G = -9.81
C_p = 3985
Res=5000
Ly=2000e3
Lx=4000e3 #Full domain = 4000km otherwise 1000km
H=4000 # Diffusion = 3800m, Sponge = 2985m
nz=40 # Diffusion = 24 level, Sponge= 30 levels
f = -1.000000E-04
Beta = 1.000000E-11
tref = 4.0
alpha = 2.000000E-04
G = -9.81
rho_0 = 999.8
dz = Zp[0:len(Zp)-1]-Zp[1:len(Zp)]
pi = math.pi
```

```python
filename = path+'/3/Tav.nc'
file2read = netCDF4.Dataset(filename,'r')
V = file2read.variables["VVEL"][:]
Vtave = np.mean(V,axis = 0)
Vtave = Vtave *Vmask # Fraction of cell open
Vzone = np.sum(Vtave*5000,axis = 2)
# Got rid of for loop here (much quicker!!)
psi = np.cumsum(-(Vzone*Zmatrix)[::-1,:],axis=0)[::-1]
npad = ((0,1), (0,0))
# Pad with zeros at bottom
psi = np.pad(psi, pad_width=npad, mode='constant', constant_values=0)
y = Yc/1000
Psi = psi/10**6  # Convert to Sv
```

```python
nx = np.round(Lx/Res)
ny = np.round(Ly/Res)
dx = np.ones(nx)*Res
dy = np.ones(ny)*Res
x = (np.cumsum(dx)-dx/2)-Lx/2
y = (np.cumsum(dy)-dy/2)-Ly/2
[Y, Z1, X] = np.meshgrid(y,Z,x)
tau = 0.2*((np.sin((Y+Ly/2)*(pi/Ly)))) 
psi = - np.apply_along_axis(np.divide,1,tau,rho_0*(f+(Beta*np.cumsum(dy))))*Tmask
psi = np.sum(5000*psi,axis=2)/10**6
npad = ((0,2), (0,0))
psi = np.pad(psi[0:-1,:], pad_width=npad, mode='constant', constant_values=0)
```

```python
fig = plt.figure(figsize=(12.5,6.5))
Psi_levs = np.arange(0.05, 17.05, 1.)
Psi_ticks =np.arange(0, 17., 1.)
ax = fig.add_subplot(1, 2, 1)
p=ax.contourf(Yc/1000,Zp/1000,psi,Psi_levs,cmap=plt.cm.Reds)
ax.set_title('a) $\overline{\psi}$', fontsize=30)
ax.set_xlabel('Distance (km)', fontsize=30)
ax.set_ylabel('Depth (km)', fontsize=30)
ax.tick_params(labelsize=20)
ax = fig.add_subplot(1, 2, 2)
ax.contourf(Yp/1000,Zp/1000,Psi,Psi_levs,cmap=plt.cm.Reds)
ax.set_title('b) $\overline{\psi}$', fontsize=30)
ax.set_xlabel('Distance (km)', fontsize=30)
ax.set_ylabel('Depth (km)', fontsize=30)
ax.tick_params(labelsize=20)
#cbar.set_label('$\psi \,\, (sv)$', fontsize=20)
plt.tight_layout()
cax = fig.add_axes([1, 0.1, 0.03, 0.8])
cbar=fig.colorbar(p, cax=cax, ticks=Psi_ticks)
cbar.ax.set_ylabel('$\psi$ (Sv)', fontsize=30)
cbar.ax.tick_params(labelsize=20) 
```
